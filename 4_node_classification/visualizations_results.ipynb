# Visualizations of results

Finally, we visualize the results of our baselines and GNNs in terms of test accuracies and best test accuracies, respectively. For this purpose, we use boxplots, which show the location and dispersion of the accuracies particularly well.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg', 'pdf')

#pd.set_option('display.max_rows', None)
#pd.set_option('display.max_columns', None)
```

## Define helper function

```python
def filter_best_hyperparameters(df, variable):
    """
    Filter the results data frame for the best hyperparameter combination(s).
    """
    return df[df[variable] == df[variable].max()]
```

## Load results

```python
# Create results folder
route0 = "./results"

if not os.path.exists(route0):
    os.mkdir(route0)
```

```python
lrc_results = pd.read_pickle(f"{route0}/lrc_results.pkl") 
rfc_results = pd.read_pickle(f"{route0}/rfc_results.pkl") 
svc_results = pd.read_pickle(f"{route0}/svc_results.pkl") 
mlp_results = pd.read_pickle(f"{route0}/mlp_results.pkl") 
gcn_results = pd.read_pickle(f"{route0}/gcn_results.pkl") 
cheb_results = pd.read_pickle(f"{route0}/cheb_results.pkl") 
gat_results = pd.read_pickle(f"{route0}/gat_results.pkl") 
```

In order to plot the accuracies for the best hyperparameter combination, we must first determine what index the row containing the test accuracies or best test accuracies of the best hyperparameter combination has.

### Logistic regression

```python
filter_best_hyperparameters(lrc_results, 'mean_val_acc')
```

The row with index 4 contains the test accuracies of the best hyperparameter combination.

### Random forest classifier

```python
filter_best_hyperparameters(rfc_results, 'mean_val_acc')
```

The row with index 335 contains the test accuracies of the best hyperparameter combination.

### Support vector classifier

```python
filter_best_hyperparameters(svc_results, 'mean_val_acc')
```

The rows with indices 466 and 1053 contain the test accuracies of the best hyperparameter combinations. We choose row 466.

### Multilayer perceptron

```python
filter_best_hyperparameters(mlp_results, 'mean_best_val_acc')
```

The rows with indices 472 and 473 contain the best test accuracies of the best hyperparameter combinations. We choose row 472.

### Graph convolutional network

```python
filter_best_hyperparameters(gcn_results, 'mean_best_val_acc')
```

The row with index 586 contains the best test accuracies of the best hyperparameter combination.

### GNN with Chebyshev convolution

```python
filter_best_hyperparameters(cheb_results, 'mean_best_val_acc')
```

The row with index 307 contains the best test accuracies of the best hyperparameter combination.

### Graph attention network

```python
filter_best_hyperparameters(gat_results, 'mean_best_val_acc')
```

The row with index 47 contains the best test accuracies of the best hyperparameter combination.

```python
lrc_best_test_accs = list(lrc_results['test_accs'].iloc[4].values())
rfc_best_test_accs = list(rfc_results['test_accs'].iloc[335].values()) 
svc_best_test_accs = list(svc_results['test_accs'].iloc[466].values())
mlp_best_test_accs = list(mlp_results['best_test_accs'].iloc[472].values())
gcn_best_test_accs = list(gcn_results['best_test_accs'].iloc[586].values())
cheb_best_test_accs = list(cheb_results['best_test_accs'].iloc[307].values())
gat_best_test_accs = list(gat_results['best_test_accs'].iloc[47].values())
```

## Boxplots

We first create the boxplot for the baselines and subsequently for the GNNs.

### Baselines

```python
# create data frame for boxplot
LRC = ['LRC'] * 10
RFC = ['RFC'] * 10
SVC = ['SVC'] * 10
MLP = ['MLP'] * 10
ACC = ['ACC'] * 10

boxplot_df_baselines = pd.concat([pd.DataFrame(list(zip(LRC, ACC, lrc_best_test_accs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(RFC, ACC, rfc_best_test_accs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(SVC, ACC, svc_best_test_accs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(MLP, ACC, mlp_best_test_accs)), columns = ['heuristic', 'metric', 'value'])],
                        ignore_index = True)

# create boxplot
plt.figure(figsize=(10,6.7))
sns.set(font_scale = 1)
sns.set_style("whitegrid")
fig = sns.boxplot(x="heuristic", y="value", hue="metric", data=boxplot_df_baselines, palette="Set3")
labels = fig.set(xlabel ="Node Classification Baseline", ylabel = "Score")

#fig.figure.savefig("./plots/node_classification_boxplot_baselines.svg", bbox_inches='tight') 
```

### GNNs

We first plot the accuracies of all baselines together to see if the boxplot is helpful.

```python
# create data frame for boxplot
GCN = ['GCN'] * 10
CHEB = ['CHEB'] * 10
GAT = ['GAT'] * 10
ACC = ['ACC'] * 10

boxplot_df_gnns = pd.concat([pd.DataFrame(list(zip(GCN, ACC, gcn_best_test_accs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(CHEB, ACC, cheb_best_test_accs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(GAT, ACC, gat_best_test_accs)), columns = ['heuristic', 'metric', 'value'])],
                        ignore_index = True)

# create boxplot
plt.figure(figsize=(10,6.7))
sns.set(font_scale = 1)
sns.set_style("whitegrid")
fig = sns.boxplot(x="heuristic", y="value", hue="metric", data=boxplot_df_gnns, palette="Set3")
labels = fig.set(xlabel ="Node Classification GNN", ylabel = "Score")
```

Due to the large differences in the accuracies, we can see very little especially for the GNN with Chebyshev convolution. 

Therefore, it makes more sense to create the boxplots of the three GNNs separately to allow an individual scaling of the y-axis.

#### GCN

```python
# create data frame for boxplot
GCN = ['GCN'] * 10
ACC = ['ACC'] * 10

boxplot_df_gcn = pd.concat([pd.DataFrame(list(zip(GCN, ACC, gcn_best_test_accs)), columns = ['heuristic', 'metric', 'value'])
                        ],
                        ignore_index = True)

# create boxplot
plt.figure(figsize=(3.3,6.7))
sns.set(font_scale = 1)
sns.set_style("whitegrid")
fig = sns.boxplot(x="heuristic", y="value", hue="metric", data=boxplot_df_gcn, palette="Set3")
labels = fig.set(xlabel ="Node Classification GNN", ylabel = "Score")

#fig.figure.savefig("./plots/node_classification_boxplot_gcn.svg", bbox_inches='tight') 
```
