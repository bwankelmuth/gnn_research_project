# Preparation of Svenska Stader Data

### svenska-stader.csv
This notebook prepares the locality, municipality and county data from `svenska-stader.csv`.

```python
# Load required packages
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

## Load data

```python
# Read in data frame
svenska_stader = pd.read_csv("../../0_raw_data/public_data/svenska-stader.csv")

# Look at entire data frame
svenska_stader
```

## Preparatory steps

```python
# make headers lower case
svenska_stader = svenska_stader.rename(columns = {'Locality': 'locality', 'Municipality': 'municipality', 'County': 'county',
                                                 'Latitude': 'latitude', 'Longitude': 'longitude'})

# delete county, latitude, longitude
svenska_stader.drop(['county', 'latitude', 'longitude'], axis = 1, inplace = True)
```

```python
# Save the prepared data frame
route0 = "../processed_data"

if not os.path.exists(route0):
    os.mkdir(route0)
    
print("saving file corresponding to svenska_stader.pkl")
svenska_stader.to_pickle(f"{route0}/svenska_stader.pkl")
pd.read_pickle(f"{route0}/svenska_stader.pkl")
```
