# Visualizations of results

Finally, we visualize the results of our baselines and GNNs in terms of test APs and AUCs as well as best test APs and AUCs, respectively. For this purpose, we use boxplots, which show the location and dispersion of the APs and AUCs particularly well.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg', 'pdf')
```

## Define helper function

```python
def filter_best_hyperparameters(df, variable):
    """
    Filter the results data frame for the best hyperparameter combination(s).
    """
    return df[df[variable] == df[variable].max()]
```

## Boxplots

We first create the boxplot for the heuristics, then for the latent-feature methods and subsequently for the GNNs.

### Heuristics 

```python
# create data frame for boxplot
CN = ['CN'] * 10
JC = ['JC'] * 10
PA = ['PA'] * 10
AA = ['AA'] * 10
RA = ['RA'] * 10
KA = ['KA'] * 10
AUC = ['AUC'] * 10
AP = ['AP'] * 10

boxplot_df_heuristics = pd.concat([pd.DataFrame(list(zip(CN, AUC, cn_test_aucs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(CN, AP, cn_test_aps)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(JC, AUC, jc_test_aucs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(JC, AP, jc_test_aps)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(PA, AUC, pa_test_aucs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(PA, AP, pa_test_aps)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(AA, AUC, aa_test_aucs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(AA, AP, aa_test_aps)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(RA, AUC, ra_test_aucs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(RA, AP, ra_test_aps)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(KA, AUC, ka_test_aucs)), columns = ['heuristic', 'metric', 'value']), 
                        pd.DataFrame(list(zip(KA, AP, ka_test_aps)), columns = ['heuristic', 'metric', 'value'])],
                        ignore_index = True)

# create boxplot
plt.figure(figsize=(10,6.7))
sns.set(font_scale = 1)
sns.set_style("whitegrid")
fig = sns.boxplot(x="heuristic", y="value", hue="metric", data=boxplot_df_heuristics, palette="Set3")
labels = fig.set(xlabel ="Link Prediction Heuristic", ylabel = "Score")

#fig.figure.savefig("./plots/link_prediction_boxplot_heuristics.svg", bbox_inches='tight') 
```
